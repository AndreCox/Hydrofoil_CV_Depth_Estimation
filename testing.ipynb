```python
import matplotlib as plt
from matplotlib.pyplot import imshow, show
import h5py
```

```python
hdf5_path = "./output/1.hdf5"
```

```python
 
with h5py.File(hdf5_path, "r") as f:
    for name, ds in f.items():
        if name == "ride_height":
            print(f"ride_height: {ds[:]}")
        if hasattr(ds, "shape") and ds.ndim >= 2:
            imshow(ds[:])
            show()
```
